Write a script that prints system path and Python version

```python
import sys
import os

# Print Python version
print("Python Version:")
print(sys.version)

# Print system path
print("\nSystem Path:")
for path in sys.path:
    print(path)
```

Explore `os` and `sys` modules to list current directory files.

```python
import os
import sys

# Show current working directory
print(f"Current Working Directory:\n{os.getcwd()}")

# List all files and folders in the current directory
print("\nFiles and Folders:")
for item in os.listdir():
    print(item)

# Show Python system path
print("\nPython sys.path:")
for path in sys.path:
    print(path)

```

Write a short script printing memory usage (psutil if available)

```python
try:
    import psutil

    # Get memory info
    mem = psutil.virtual_memory()
    print("Memory Usage:")
    print(f"Total: {mem.total / (1024 ** 3):.2f} GB")
    print(f"Available: {mem.available / (1024 ** 3):.2f} GB")
    print(f"Used: {mem.used / (1024 ** 3):.2f} GB")
    print(f"Percentage: {mem.percent}%")

except ImportError:
    print("psutil module not found. Please install it using 'pip install psutil'.")

```
